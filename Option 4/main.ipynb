```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from scipy.spatial import Voronoi, voronoi_plot_2d
```

### Axis and Graphing

```python
data = pd.read_csv('Fish.csv', sep=",")
print(data)
```

### Spliting and fitting the data

```python
X = list(zip(data["Length3"], data["Height"]))
x1 = data["Length3"]
y1 = data["Height"]

kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(X)
kcenters = kmeans.cluster_centers_
colorized = kmeans.predict(X) # colores data based on clustering place
print(X, kcenters)
```

### Graphing

```python
plt.scatter(x1, y1, c=colorized)
plt.scatter(kcenters[0][0], kcenters[0][1], c='red')
plt.annotate(f'({round(kcenters[0][0])}, {round(kcenters[0][1])})', (kcenters[0][0], kcenters[0][1]), textcoords="offset points", xytext=(0,-20), ha='center', c='red', fontsize=15)
plt.scatter(kcenters[1][0], kcenters[1][1], c='red')
plt.annotate(f'({round(kcenters[1][0])}, {round(kcenters[1][1])})', (kcenters[1][0], kcenters[1][1]), textcoords="offset points", xytext=(0,-20), ha='center', c='black', fontsize=15)
plt.scatter(kcenters[2][0], kcenters[2][1], c='red')
plt.annotate(f'({round(kcenters[2][0])}, {round(kcenters[2][1])})', (kcenters[2][0], kcenters[2][1]), textcoords="offset points", xytext=(0,-20), ha='center', c='blue', fontsize=15)
plt.title('Length of fish used to predict weight (KMeans)', fontsize=20)
plt.xlabel('Length of Fish', fontsize=15)
plt.ylabel('height of Fish', fontsize=15)
plt.grid()
```

### That Was Bad, So Lets do Somthing Different,
How about we use Gaussian Mixture to combat the mixture of groups

### Gaussian Algorithem:
### $$ pdf(x)=\sum_{k=1}^K w_j \times N(X|\phi) $$

```python
gauss = GaussianMixture(n_components=5).fit_predict(X)
print(gauss)
```

### Graphing again

```python
plt.scatter(x1, y1, c=gauss)
plt.title('Length of fish used to predict weight (Gaussian)', fontsize=20)
plt.xlabel('Length of Fish', fontsize=15)
plt.ylabel('height of Fish', fontsize=15)
plt.grid()
```
